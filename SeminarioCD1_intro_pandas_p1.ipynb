# **Series y DataFrames**

*   Series:

Son estructuras unidimensionales que pueden almacenar una sola columna de cualquier tipo de datos, como números, cadenas o booleanos. Se utilizan a menudo para representar una columna de datos en un dataframe.

*   DataFrames:

Son estructuras bidimensionales que pueden almacenar varias columnas de diferentes tipos de datos. Están compuestas por filas y columnas, y cada columna es una serie. Los dataframes permiten destacar las relaciones entre las distintas variables de la serie de datos.

```python
personas=["Ana", "José", "Luis", "María"] #lista de python
```

```python
serie=pd.Series(personas) #convertir la lista de personas en una Serie de pandas
print(serie)
```

```python
empleados=pd.DataFrame([{"Nombre": "Ana", "Apellido": "Suarez", "Edad":35, "Contrato_fijo": 0},
                         {"Nombre": "José", "Apellido": "López", "Edad":28, "Contrato_fijo": 1}],
                        columns=["Nombre", "Apellido", "Edad", "Contrato_fijo"])

print(empleados)
```

**Count** - nos muestra el conteo de registros

**unique** - nos dice cuantos tipos de registros hay, en este caso 2 (0s y 1s)

**top** - el más frecuente

**freq** - frecuencia máxima

```python
var_cat=['blue', 'three_g', 'four_g', 'wifi', 'touch_screen'] #lista de variables que deberían tener categorías
for variable in var_cat: #para cada elemento de la lista var_cat
    data[variable] = data[variable].astype('category') #convertir la variable a categorías
```
