```python
from scipy import stats
```

```python
p75ages = stats.scoreatpercentile(data["age"], 75)
print("Percentil 75 = {}".format(p75ages))

p10ages = stats.scoreatpercentile(data["age"], 10)

print("Percentil 10 = {}".format(p10ages))

q2ages = stats.scoreatpercentile(data["age"], 50)

print("Cuartil 2 = {}".format(q2ages))
```
